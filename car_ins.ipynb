filling missing values

```python
df['CREDIT_SCORE'].unique()
```

```python
df['ANNUAL_MILEAGE'].unique()
```

```python
df['CREDIT_SCORE'].fillna(df['CREDIT_SCORE'].mode()[0],inplace=True)
```

```python
df['ANNUAL_MILEAGE'].fillna(df['ANNUAL_MILEAGE'].mode()[0],inplace=True)
```

```python
df.isna().sum()
```

```python
#dropping unwanted column
df.drop(['ID'],axis=1,inplace=True)
```

```python
df
```

```python
#encode categorical features using labelencoder
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
for i in df.columns:
    if df[i].dtypes=='object':
        df[i]=encoder.fit_transform(df[i])
```

```python
df
```

```python
#splitiing dataset
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x.ndim
```

```python
y.ndim
```

```python
#splitting testing and training dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=.30)
```

```python
y_train
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
#model training
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train,y_train)
#predicting 
y_pred=model.predict(x_test)
y_pred
```

Evaluation

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay,classification_report,accuracy_score
score=accuracy_score(y_test,y_pred)
score
```

```python
cd=confusion_matrix(y_test,y_pred)
cd
```

```python
labels=['No claim','Made a claim']
cmd=ConfusionMatrixDisplay(cd,display_labels=labels)
cmd.plot()
```

```python
report=classification_report(y_test,y_pred)
print(report)
```
